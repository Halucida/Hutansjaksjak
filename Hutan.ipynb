```python
from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.metrics.pairwise import linear_kernel
```

```python
tif = TfidfVectorizer(stop_words="english")

tfidf = tif.fit_transform(film_df["Genre"])
```

```python
tfidf.shape
```

```python
film_similarity = linear_kernel(tfidf, tfidf)

len(film_similarity), len(film_similarity[0])
```

```python
ind_tit = lambda i : film_df.iloc[i, 1]

tit_ind = lambda t : film_df[film_df["Title"] == t].index[0]
```

```python
ml_list = list(enumerate(film_similarity[int(tit_ind("Toy Story"))]))

ml_list = sorted(ml_list, key = lambda x:x[1], reverse=True)
```

```python
def recommender(title, threshold):


  m_list = list(enumerate(film_similarity[int(tit_ind(title))]))

  m_list = sorted(m_list, key = lambda i : i[1], reverse=True)

  m_list = m_list[:threshold]

  m_list = list(filter(lambda i : i[0] != tit_ind(title), m_list))

  m_list = [ind_tit(m_list[i][0]) for i in range(len(m_list))]

  return film_df[film_df["Title"].isin(m_list)]

recommender("Jumanji", 5)
```
